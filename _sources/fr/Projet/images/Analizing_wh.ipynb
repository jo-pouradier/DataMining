# Resizing all images to the fittest size in 16:9 ratio

```python
# getting all images
import os
from PIL import Image
import numpy as np
themes = ["Exoplanet", "Moto","Pokemon2", "Voitures"]
width_possible = [120, 640, 960, 1280, 1600, 1920, 2880]
height_possible = [x*9/16 for x in width_possible]
height_possible[0] = 120

#making correspond width and height
sizes = list(zip(width_possible, height_possible))
print(sizes)
sizes = np.array(sizes)
```

```python
def get_new_size(img_path):
    img = Image.open(img_path)
    size = img.size
    print(size)
    size = np.array(size)
    if size in sizes:
        print("size is ok")
        return
    # get the closest height np.abs(np.array(height_possible) - height).argmin()
    sizes_distance = np.sqrt(np.sum((sizes-size)**2, axis=1))
    index_new_size = np.where(sizes_distance == np.amin(sizes_distance))
    new_size = sizes[index_new_size]
    print(new_size[0])
    # resize image
    img = img.resize((int(new_size[0][0]), int(new_size[0][1])))
    # save image
    img.mode = "RGB"
    img.save(img_path)

get_new_size("Exoplanet/Mu_Arae_b.jpg")
```

```python

for theme in themes:
    print(theme)
    for image in os.listdir(theme):
        if image.endswith(".jpg") or image.endswith(".png"):
                img_path = os.path.join(theme, image)
                get_new_size(img_path)
```
