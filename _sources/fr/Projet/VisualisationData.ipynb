# Visualisation de notre Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#import data
df = pd.read_json('./images/ExifDatatest2.json')
df.drop(['image', 'format'], axis=1, inplace=True)
df
```

```python
# visualise data with seaborn
sns.pairplot(df, height=2.5, hue='theme', y_vars='theme', x_vars=['width', 'height', 'color_0', 'color_1', 'color_2', 'color_3', 'color_4', 'red', 'green', 'blue'], kind='hist')
```
