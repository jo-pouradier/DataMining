# Visualisation de notre Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#import data
df = pd.read_json('./images/ExifDatatest2.json')
df.drop(['image', 'format', 'orientation'], axis=1, inplace=True)
df
```

## image par theme

```python
## number of element per theme
df['theme'].value_counts().plot(kind='bar', title='Number of element per theme', figsize=(10,5), color=['red', 'green', 'blue','purple'])
```

# image pour chaque type

```python
## number of element per value of each columns
plt.figure(figsize=(25,17))
# set title
plt.suptitle('Number of element per value of each columns', fontsize=30)
# add space between subplots
plt.subplots_adjust(hspace=0.3, wspace=0.1)

# loop over columns
for col in df.columns:
    plt.subplot(4,3,df.columns.get_loc(col)+1)
    df[col].value_counts().plot(kind='bar', title=col, color=['red', 'green', 'blue','purple'], rot=45)


```
