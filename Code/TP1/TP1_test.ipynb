# Exercice 4

```python
import urllib.request
import json
import pandas as pd

url = "https://query.wikidata.org/sparql?query=SELECT%20%3FlanguageLabel%20(YEAR(%3Finception)%20as%20%3Fyear)%0AWHERE%0A%7B%0A%20%20%23instances%20of%20programming%20language%0A%20%20%3Flanguage%20wdt%3AP31%20wd%3AQ9143%3B%0A%20%20%20%20%20%20%20%20%20%20%20%20wdt%3AP571%20%3Finception%3B%0A%20%20%20%20%20%20%20%20%20%20%20%20rdfs%3Alabel%20%3FlanguageLabel.%0A%20%20FILTER(lang(%3FlanguageLabel)%20%3D%20%22en%22)%0A%7D%0AORDER%20BY%20%3Fyear%0ALIMIT%20100&format=json"
response = urllib.request.urlopen(url)
responsedata = json.loads(response.read().decode("utf-8"))

array = []

for data in responsedata["results"]["bindings"]:
    array.append([data["year"]["value"], data["languageLabel"]["value"]])

dataframe = pd.DataFrame(array, columns=["year", "languageLabel"])
dataframe = dataframe.astype(dtype={"year": "<i4", "languageLabel": "<U200"})
print(dataframe)
```

```python
dataframe["year"].describe()
```

```python
dataframe["languageLabel"].describe()
```

```python
from pandas.io.json import json_normalize
import pandas as pd
import json

jsondata = json.load(open("../../data/plparadigm.json"))
array = []

for data in jsondata:
    array.append([data["year"], data["languageLabel"], data["paradigmLabel"]])

dataframe = pd.DataFrame(array, columns=["year", "languageLabel", "paradigmLabel"])
dataframe = dataframe.astype(
    dtype={"year": "int64", "languageLabel": "<U200", "paradigmLabel": "<U200"}
)

grouped = dataframe.groupby(["year", "paradigmLabel"]).agg(["count"])
grouped
```

```python
grouped = dataframe.groupby(["paradigmLabel", "year"]).agg(["count"])
grouped
```

## Utilisation des data wikipedia json

```python
jsondata = json.load(open("queryfull.json"))

array = []
for data in jsondata:
    array.append([data["countryLabel"], data["year"], data["population"]])

dataframe = pd.DataFrame(array, columns=["countryLabel", "year", "population"])
dataframe = dataframe.astype(
    dtype={"countryLabel": "<U200", "year": "int16", "population": "int64"}
)

dataframe["countryLabel"].describe()
dataframe.groupby(["countryLabel", "year", "population"])
dataframe
#population_asc_country = dataframe.groupby(["countryLabel", "year", "population"])
#population_asc_country
```

```python
df = dataframe.sort_values(["countryLabel","year"])
print(df)
```

```python
df = dataframe.loc[dataframe.groupby("countryLabel")["year"].idxmax()]
df
```

```python
df = dataframe.loc[dataframe.groupby("countryLabel")["year"].idxmax()].sort_values("population" , ascending=False)
df
```

```python
df = dataframe.loc[dataframe.groupby("countryLabel")["year"].idxmax()].sort_values("population" , ascending=True)
df
```

### Articles scientifiques

```python
jsondata = json.load(open("queryScience.json"))

array = []
for data in jsondata:
    array.append([data["title"], data["subjectLabel"], data["year"]])

dataframe = pd.DataFrame(array, columns=["title", "subjectLabel", "year"])
dataframe = dataframe.astype(
    dtype={"title": "<U400", "subjectLabel": "<U100", "year": "int16"}
)

dataframe["title"].describe()
dataframe.groupby(["title", "subjectLabel", "year"])
dataframe
```

```python
df = dataframe.groupby(["subjectLabel", "year"]).agg("count")
df
```

```python
df = dataframe.loc[dataframe.groupby("subjectLabel")["year"].idxmax()].sort_values("year" , ascending=True)
df
```
