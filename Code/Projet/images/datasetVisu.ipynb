```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Read the df
df = pd.read_json('Exifdatatest2.json')
df.drop('orientation', axis=1, inplace=True)
df.drop('image', axis=1, inplace=True)
df.drop('format', axis=1, inplace=True)
df.head()
```

## Labelisation

```python
df['width'] = df['width'].astype('category').cat.codes
df['height'] = df['height'].astype('category').cat.codes
df['color_0'] = df['color_0'].astype('category').cat.codes
df['color_1'] = df['color_1'].astype('category').cat.codes
df['color_2'] = df['color_2'].astype('category').cat.codes
df['color_3'] = df['color_3'].astype('category').cat.codes
df['color_4'] = df['color_4'].astype('category').cat.codes
df['red'] = df['red'].astype('category').cat.codes
df['green'] = df['green'].astype('category').cat.codes
df['blue'] = df['blue'].astype('category').cat.codes
df.head()
```

```python
# Plot the data
sns.pairplot(df, hue='theme', diag_kind='kde', height=2.5)
```
