# Transformation de PNG a JPG

```python

from PIL import Image
import PIL
import os
  
dir = './images/Pokemon'
files = os.listdir(dir)
files.remove('.DS_Store')
#open image in png formats
for file in files:
    print(file, end='\r')
    try:
        img_png = Image.open(dir+ '/' + file, formats=['png'])
        img_jpg = img_png.convert('RGB')
        img_jpg.save(dir+'2/' + file[:-4] + '.jpg')
    except PIL.UnidentifiedImageError:
        print('\nerror\n')
#The image object is used to save the image in jpg format
```
